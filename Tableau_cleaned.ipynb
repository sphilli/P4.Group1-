```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
df = pd.read_csv("Resources/original_data.csv")

# Review the DataFrame
df.head()
```

```python
df.info()
```

```python
# Conditionally update values in the 'Gender'column
new_value_male = 'Male'
new_value_female = 'Female'

# Using .loc with boolean mask
df.loc[df['Gender'] == 0, 'Gender'] = new_value_female
df.loc[df['Gender'] == 1, 'Gender'] = new_value_male

df.head()
```

```python
# Conditionally update values in the 'Own_car'column
car_owner = 'Yes'
car_nonowner = 'No'

# Using .loc with boolean mask
df.loc[df['Own_car'] == 0, 'Own_car'] = car_nonowner
df.loc[df['Own_car'] == 1, 'Own_car'] = car_owner

df.head()
```

```python
# Conditionally update values in the 'Own_property'column
home_owner = 'Yes'
home_nonowner = 'No'

# Using .loc with boolean mask
df.loc[df['Own_property'] == 0, 'Own_property'] = home_nonowner
df.loc[df['Own_property'] == 1, 'Own_property'] = home_owner

df.head()
```

```python
# Conditionally update values in the 'Unemployed'column
Unemployed = 'Yes'
Employed = 'No'

# Using .loc with boolean mask
df.loc[df['Unemployed'] == 0, 'Unemployed'] = Employed
df.loc[df['Unemployed'] == 1, 'Unemployed'] = Unemployed

df.head()
```

```python
# Conditionally update values in the 'Target'column
risk_low = 'Low'
risk_high = 'High'

# Using .loc with boolean mask
df.loc[df['Target'] == 0, 'Target'] = risk_low
df.loc[df['Target'] == 1, 'Target'] = risk_high

df.head()
```

```python
# Drop unwanted columns
columns_to_drop = ['Work_phone', 'Phone' , 'Email']
df.drop(columns=columns_to_drop, inplace=True)
```
